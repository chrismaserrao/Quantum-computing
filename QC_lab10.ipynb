#Lab 10
simulate the Diffie–Hellman key exchange in Python and demonstrate how a Man-in-the-Middle attack can intercept and manipulate the shared keys.

Implement standard Diffie–Hellman key exchange using a prime number and generator.

Create three entities: Alice, Bob, and Tom (the attacker).

Make Tom intercept and replace the public keys exchanged between Alice and Bob.

Show that:
Alice shares a secret key with Tom
Bob shares a different secret key with Tom
Alice and Bob do NOT share the same key




Alice ↔ Tom share the same secret key

Bob ↔ Tom share another secret key

Alice and Bob DO NOT share a key

Tom intercepts and replaces each public key with his own.

```python
import random

# ------------------------------
# Global Prime (p) and Generator (g)
# ------------------------------
p = 23               # Small prime for demonstration
g = 5                # Generator

print("Public parameters:")
print("Prime p =", p)
print("Generator g =", g)
print("-" * 40)

# ------------------------------
# Step 1: Alice and Bob Choose Private Keys
# ------------------------------
alice_private = random.randint(2, p-2)
bob_private   = random.randint(2, p-2)

print("Alice Private Key:", alice_private)
print("Bob Private Key:  ", bob_private)
print("-" * 40)

# ------------------------------
# Step 2: Alice and Bob Compute Public Keys
# ------------------------------
alice_public = pow(g, alice_private, p)
bob_public   = pow(g, bob_private, p)

print("Alice Public Key (original):", alice_public)
print("Bob Public Key   (original):", bob_public)
print("-" * 40)

# ------------------------------
# Tom the Attacker (MitM)
# ------------------------------
tom_private = random.randint(2, p-2)
tom_public  = pow(g, tom_private, p)

print("Tom Private Key:", tom_private)
print("Tom Public Key :", tom_public)
print("-" * 40)

# Tom intercepts and modifies
alice_public_sent_to_bob = tom_public      # Replaces Alice's public key
bob_public_sent_to_alice = tom_public      # Replaces Bob's public key

print("Alice sends → Tom → Bob:", alice_public_sent_to_bob)
print("Bob sends   → Tom → Alice:", bob_public_sent_to_alice)
print("-" * 40)

# ------------------------------
# Step 3: Each party computes the "shared" key
# ------------------------------
# Alice computes key using fake public key from Tom
alice_shared_key = pow(bob_public_sent_to_alice, alice_private, p)

# Bob computes key using fake public key from Tom
bob_shared_key   = pow(alice_public_sent_to_bob, bob_private,   p)

# Tom computes both keys
tom_shared_with_alice = pow(alice_public, tom_private, p)
tom_shared_with_bob   = pow(bob_public,   tom_private, p)

# ------------------------------
# Show Results
# ------------------------------
print("Shared Key (Alice ↔ Tom):", alice_shared_key)
print("Shared Key (Tom   ↔ Alice):", tom_shared_with_alice)
print()

print("Shared Key (Bob   ↔ Tom):", bob_shared_key)
print("Shared Key (Tom   ↔ Bob):", tom_shared_with_bob)
print()

print("Do Alice and Bob share the same key?")
print("Alice-Bob key comparison:", alice_shared_key == bob_shared_key)
```
